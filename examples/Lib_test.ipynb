```python
import time
import math
```

```python
# move front en checkant distance
# quand devant obstacle il s'arrete
# rot 270 trigo
# avance C temps/distance
# step left
# lecture carte couleur et met à jour ses leds
# tourne 90 degres clock
# slide right C temps/distance (same as etape précendente)
# avance jusqua ce que l'obstacle derriere lui est à X distance


def test_robot_ilo():

    # Step 1: check the initial distance
    my_first_distance = my_ilo.get_distance_front()


    # Step 2: Move forward until the robot is 10 cm from the obstacle
    while (my_ilo.get_distance_front() > 10):
        my_ilo.move("front", 10, 50) # Move forward (10% speed, 50% acceleration)
        time.sleep(0.1)
    my_ilo.pause() # Stop the robot

    # Step 3: Rotate the robot to the right
    my_ilo.step("rot_clock")  #True at default, mean we are waiting the end the displacement

    # Step 4: Move forward slightly and turn on green LEDs
    my_ilo.set_led_color(25,200,25)
    my_ilo.step("front", 0.5, False)
    
    # Step 5: Play a “wave” LED animation
    my_ilo.set_led_anim("wave", 1)

    # Step 6: Rotate twice by 45° and print status each time
    my_ilo.rotation(45)
    print("end of first 45 deg")
    my_ilo.rotation(45)
    print("end of second 45 deg")
    
    # Step 7: Read the color under the robot and set LEDs to that color
    my_ilo.set_led_captor(200)
    time.sleep(1)
    color_under_ilo =  my_ilo.get_color_card("rgb")
    r = color_under_ilo[0]
    g = color_under_ilo[1]
    b = color_under_ilo[2]
    my_ilo.set_led_color(r,g,b)
    
    # Step 8: Move 0.5 m to the right
    my_ilo.step("right", 0.5)
    
    # Step 9: Move forward until back distance equals the initial front distance, then stop
    while (my_ilo.get_distance_back() < my_first_distance):
        my_ilo.move("front", 10, 50)
    my_ilo.stop()
        
test_robot_ilo()
    
```
