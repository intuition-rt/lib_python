```python
# move front en checkant distance
# quand devant obstacle il s'arrete
# rot 270 trigo
# avance C temps/distance
# step left
# lecture carte couleur et met à jour ses leds
# tourne 90 degres clock
# slide right C temps/distance (same as etape précendente)
# avance jusqua ce que l'obstacle derriere lui est à X distance


def test_robot_ilo():

    #step1 1: check the initial distance
    my_first_distance = my_ilo.get_distance_front()


    #step 2: go to the obstacle unitl the last moment
    while (my_ilo.get_distance_front() > 10):
        my_ilo.move("front", 10, 50) # speed at 10% et acceleration at 50% 
        time.sleep(0.1)
    my_ilo.pause() #stop the robot at keep the 

    #step 3: turn ilo on the right
    my_ilo.step("rot_clock")  #True at default, mean we are waitng the end the displacement

    #step 3: move ilo forware and use the LED (change color and play one animation)
    my_ilo.set_led_color(25,200,25)
    my_ilo.step("front", 0.5, False)
    
    #step 4: make animation with LED
    my_ilo.set_led_anim("wave", 1)

    #step 5: make rotation of ilo
    my_ilo.rotation(45)
    print("end of first 45 deg")
    my_ilo.rotation(45)
    print("end of second 45 deg")
    
    #step 6: check the color under the robot and display it on the LED
    my_ilo.set_led_captor(200)
    time.sleep(1)
    color_under_ilo =  my_ilo.get_color_card("rgb")
    r = color_under_ilo[0]
    g = color_under_ilo[1]
    b = color_under_ilo[2]
    my_ilo.set_led_color(r,g,b)
    
    #step 7: return front of the obstacle
    my_ilo.step("right", 0.5)
    
    #step 8: return front the obstacle
    while (my_ilo.get_distance_back() < my_first_distance):
        my_ilo.move("front", 10, 50)
    my_ilo.stop()
    
    #step 9: rotation around the led
    #my_ilo.get_angle()[2]
    

    
test_robot_ilo()
    
```
